```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('seaborn')
i1 = cv2.imread(r'C:\Users\Gaurab\Desktop\bottom_left.jpg')
i2 = cv2.imread(r'C:\Users\Gaurab\Desktop\bottom_right.jpg')
i3 = cv2.imread(r'C:\Users\Gaurab\Desktop\top_left.jpg')
i4 = cv2.imread(r'C:\Users\Gaurab\Desktop\top_right.jpg')
i5 = cv2.imread(r'C:\Users\Gaurab\Desktop\center.jpeg')
img1 = cv2.resize(i1, (200, 200))
img2 = cv2.resize(i2, (200, 200))
img3 = cv2.resize(i3, (200, 200))
img4 = cv2.resize(i4, (200, 200))
img5 = cv2.resize(i5, (100, 100))
# img5 = cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)
img1 = cv2.copyMakeBorder(img1, 5, 10, 10, 5, cv2.BORDER_CONSTANT)
img2 = cv2.copyMakeBorder(img2, 5, 10, 5, 10, cv2.BORDER_CONSTANT, 1)
img3 = cv2.copyMakeBorder(img3, 10, 5, 10, 5, cv2.BORDER_CONSTANT, 1)
img4 = cv2.copyMakeBorder(img4, 10, 5, 5, 10, cv2.BORDER_CONSTANT, 1)
img5 = cv2.copyMakeBorder(img5, 10, 10, 10, 10, cv2.BORDER_CONSTANT, 1)
np_h1 = np.hstack((img3, img4))
np_h2 = np.hstack((img1, img2))
ans = np.concatenate((np_h1, np_h2), axis = 0)
ans = cv2.cvtColor(ans,cv2.COLOR_BGR2RGB)
plt.imshow(ans)
```

```python
s_img = img5
s_img = cv2.cvtColor(s_img,cv2.COLOR_BGR2RGB)
l_img = ans
x_offset = y_offset = 150
l_img[y_offset:y_offset+s_img.shape[0], x_offset:x_offset+s_img.shape[1]] = s_img
ans = l_img
plt.axis("off")
plt.imshow(ans)
ans.shape
df = pd.DataFrame(ans.reshape(430*430, 3), columns=list('RGB'))
df.to_csv('pokemon_data.csv', index = False)
```
